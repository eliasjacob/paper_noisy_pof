```python
equiv_raw[0]

dict_equiv = {}

for line in equiv_raw:
    sigla_col = line[0][0]
    nome_col = line[0][1]
    dict_equiv[sigla_col] = {}
    dict_equiv[sigla_col]['nome'] = nome_col

    dict_equiv[sigla_col]['equivalencias'] = {}
    #print('line ', sigla_col, len(line))
    if len(line) > 1:
        for i in range(1, len(line)):
            #print('i ', i)
            dict_equiv[sigla_col]['equivalencias'][line[i][0]] = line[i][1]
        
```

```python
dict_equiv
```

```python
import re
import unidecode

def prepare_column_name(cod, txt):

    txt = txt.replace('_', ' ')    
    # remove symbols from column name
    txt = re.sub(r'[^\w\s]', '', txt)
    # remove accents from column name
    txt = unidecode.unidecode(txt)
    # remove multiple spaces from column name
    txt = re.sub(' +', ' ', txt)
    # remove spaces from column name
    txt = txt.replace(" ", "_")


    text = f"{cod}____{txt}"
    return text

df_mapped = df.copy()

for col in df_mapped.columns:
    if col in dict_equiv.keys():
        if len(dict_equiv[col]['equivalencias']) > 0:
            df_mapped[col] = df_mapped[col].astype(str)
            df_mapped[col] = df_mapped[col].replace(dict_equiv[col]['equivalencias'])
    #        print(col, dict_equiv[col]['nome'])
        
        df_mapped.rename(columns={col: prepare_column_name(col, dict_equiv[col]['nome'])}, inplace=True)

```

```python
# Do quadro 1 - 230605_cc_59_ota_21_rendimentos.pdf - Faixas de renda mensal domiciliar 

valores_originais_ipea_2020 = {
    1: 1650.50,
    2: 2471.09,
    3: 4127.41,
    4: 8254.83,
    5: 16509.66,
}


def ajustar_sm_ipea(valor_original):
    SM_2020 = 1045.00
    return valor_original / SM_2020

valores_ajustados_sm_2020 = {k: ajustar_sm_ipea(v) for k, v in valores_originais_ipea_2020.items()}
valores_ajustados_sm_2020

# Adicionei uma faixa 0 para pessoas muito muito pobres (1/2 da faixa 1)
# Adicionei uma faixa 6 para pessoas muito muito ricas (2x a faixa 5)
# Acima da faixa 6 é a faixa 7.

valores_ajustados_sm_2020[0] = valores_ajustados_sm_2020[1] / 2
valores_ajustados_sm_2020[6] = valores_ajustados_sm_2020[5] * 2

# sort the dictionary by key
valores_ajustados_sm_2020 = dict(sorted(valores_ajustados_sm_2020.items(), key=lambda item: item[0]))
valores_ajustados_sm_2020
```

```python
def check_category(x):
    for i, cat in valores_ajustados_sm_2020.items():
        if x <= cat:
            return i
    return len(valores_ajustados_sm_2020)
```
