```python
df['imc_medio_familia'] = df.groupby('idfam')['imc'].transform('mean')
df['energia_kcal_medio_familia'] = df.groupby('idfam')['ENERGIA_KCAL'].transform('mean')
df['anos_estudo_medio_familia'] = df.groupby('idfam')['ANOS_ESTUDO'].transform('mean')
df['idade_media_familia'] = df.groupby('idfam')['idade'].transform('mean')
```

```python
df.seguranca_alimentar.unique()
```

```python
df.seguranca_alimentar.value_counts()

df['pior_situacao_san'] = 'SAN'

for idfam in df.idfam.unique():
    df_tmp = df[df.idfam == idfam].copy()
    valores_san = df_tmp.seguranca_alimentar.unique()
    if 'IA grave' in valores_san:
        df.loc[df.idfam == idfam, 'pior_situacao_san'] = 'IA grave'
    elif 'IA moderada' in valores_san:
        df.loc[df.idfam == idfam, 'pior_situacao_san'] = 'IA moderada'
    elif 'IA leve' in valores_san:
        df.loc[df.idfam == idfam, 'pior_situacao_san'] = 'IA leve'
    elif 'SAN' in valores_san:
        df.loc[df.idfam == idfam, 'pior_situacao_san'] = 'SAN'

df['pior_situacao_san'].value_counts()

```

```python
import re
import unidecode

def prepare_column_name(cod, txt):

    txt = txt.replace('_', ' ')    
    # remove symbols from column name
    txt = re.sub(r'[^\w\s]', '', txt)
    # remove accents from column name
    txt = unidecode.unidecode(txt)
    # remove multiple spaces from column name
    txt = re.sub(' +', ' ', txt)
    # remove spaces from column name
    txt = txt.replace(" ", "_")


    text = f"{cod}____{txt}"
    return text

df_mapped = df.copy()

for col in df_mapped.columns:
    if col in dict_equiv.keys():
        if len(dict_equiv[col]['equivalencias']) > 0:
            df_mapped[col] = df_mapped[col].astype(str)
            df_mapped[col] = df_mapped[col].replace(dict_equiv[col]['equivalencias'])
    #        print(col, dict_equiv[col]['nome'])
        
        df_mapped.rename(columns={col: prepare_column_name(col, dict_equiv[col]['nome'])}, inplace=True)

```
